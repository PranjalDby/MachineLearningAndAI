### SHOWING ONLY "Y POINTS" WITH FMT PARAMATER MARKER|LINE|COLOR

```python
xs_new = 1 + 2.5 * np.random.randn(5)
ys_new = 2 - 1.5 * np.random.randn(5)
print("xs_new = ",xs_new)
print("ys_new = ",ys_new)
# mec :: marker edge color
# ms :: marker size
# mfc :: marker face color
plt.plot(xs_new,ys_new,'*-y',mec='r')
plt.grid(True)
plt.xlabel("x-axis",fontdict={"family":"serif","color":'violet','size':20})
plt.ylabel("y-axis")
plt.show()
```

```python
plt.subplot(1,2,1)## plotting multiple figures within the same plot
#here 1:no of rows 2:no of columns and 1:index(plot number)
plt.plot(xs_new,':r')
plt.plot(ys_new,'-.y')
# Rotate x-axis and y-axis tick labels by 45 degrees
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.show()
```

# Important Plots and Graphs in Matplotlib

Matplotlib is a powerful Python library for data visualization. Here are some of the most important plots and graphs you can create with matplotlib:

## 1. Line Plot
- **Purpose:** Visualize the relationship between two variables, typically continuous data.
- **Usage:** `plt.plot(x, y)`
- **Example:** Showing trends over time or regression relationships.

## 2. Scatter Plot
- **Purpose:** Show the relationship between two sets of data using points.
- **Usage:** `plt.scatter(x, y)`
- **Example:** Visualizing correlation or clustering in data.

## 3. Bar Plot
- **Purpose:** Compare quantities corresponding to different groups or categories.
- **Usage:** `plt.bar(categories, values)` or `plt.barh(categories, values)`
- **Example:** Comparing sales of different products.

## 4. Histogram
- **Purpose:** Display the distribution of a dataset.
- **Usage:** `plt.hist(data, bins=number_of_bins)`
- **Example:** Understanding the frequency distribution of exam scores.

## 5. Pie Chart
- **Purpose:** Show proportions of a whole.
- **Usage:** `plt.pie(sizes, labels=labels)`
- **Example:** Market share of different companies.

## 6. Box Plot
- **Purpose:** Summarize data using quartiles and identify outliers.
- **Usage:** `plt.boxplot(data)`
- **Example:** Comparing distributions between several groups.

## 7. Subplots
- **Purpose:** Display multiple plots in a single figure for comparison.
- **Usage:** `plt.subplot(rows, cols, index)`
- **Example:** Comparing different visualizations side by side.

---

**Tips:**
- Customize plots with titles, labels, legends, and colors.
- Use `plt.grid(True)` for better readability.
- Save figures using `plt.savefig('filename.png')`.

Matplotlib offers flexibility and control for creating publication-quality visualizations in Python.

```python
exam_scores = np.random.randint(0,100,size=10)

plt.hist(exam_scores,bins=4)
plt.ylabel("frequency")
plt.xlabel("MARKS")
plt.show()
```

#SUBPLOTS


```python
x = [1,2,3,4,5]
y1 = [2,3,5,7,11]
y2 = [1,4,6,8,10]
# plt.subplots() ## it will return a tuple of objects (figure and plots)
fig, axs_plots = plt.subplots(2)
#axs_plots here represent horizontal axis means (axis = 0)
axs_plots[0].plot(x,y1) # 0 represents index
axs_plots[0].set_title("First Plot")
axs_plots[1].plot(x,y2,'-r')
axs_plots[1].set_title("Second Plot")
plt.tight_layout(pad=0.8)
plt.show()
```
