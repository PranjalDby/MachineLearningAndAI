```python
import pandas as pd
import numpy as np


df = pd.read_csv('./weather_ww2.csv',low_memory=False)

df.head()

print("No of nan = ",df.isna().sum())
df = df.drop(axis=1,labels=['WindGustSpd'])
#dropping na values bz dataset is huge so no matters
print("cols = ",df.columns)
df = df.dropna(axis=1,how='all')
print("----------------------------")
print("after dropping na rows = ",df.isna().sum())

print("Replacinn na with its mean value.")
df = df.fillna(0)
```

```python
# Clean 'Snowfall' column to ensure all values are numeric
df['Snowfall'] = pd.to_numeric(df['Snowfall'], errors='coerce').fillna(0)
print(np.corrcoef(df['MinTemp'], df['Snowfall'].astype(float)))
```

```python
from tabulate import tabulate # type: ignore
## correlation matrix

corr_matr = df.corr(numeric_only=True) # karl-pearson

print(tabulate(corr_matr,headers='keys',tablefmt='grid',numalign='right',floatfmt='.2f'))
```
